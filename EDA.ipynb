```python
ax = df['Year_Introduced'].value_counts() \
  .head(10) \
  .plot(kind='bar')
ax.set_xlabel('year Introduced')
ax.set_ylabel('counts')
```

```python
bx = df['Speed_mph'].plot(kind='hist')
bx = bx.twinx()  # Create a twin Axes sharing the xaxis
df['Speed_mph'].plot(kind='kde', ax=bx, color='r')

bx.set_xlabel('mph')
bx.set_ylabel('counts')
plt.show()
```

#Asking Question About Data
###1.Location of Fastest Rollercoster(10)

```python
df.head()
```

```python
df.query('Location != "Other"') \
  .groupby('Location')['Speed_mph'] \
  .agg(['mean', 'count']) \
  .query('count >= 10') \
  .sort_values('mean')['mean'] \
  .plot(kind='barh' , figsize=(12,5))

plt.show()
```
